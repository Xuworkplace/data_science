# Grouping

```python
# Extract unique group numbers

group_numbers = CCSJ_list['Group number'].unique()
```

```python
# Create a dictionary to store dataframes for each group

grouped_dfs = {}
```

```python
# Iterate through each group number and filter the rows in the original dataframe

for group in group_numbers:
    ccsj_list = CCSJ_list[CCSJ_list['Group number'] == group]['CCSJ'].tolist()
    grouped_dfs[group] = seva_cr[seva_cr['CCSJ'].isin(ccsj_list)]
```

```python
# Separate the grouped data into three different dataframes

group_1_df = grouped_dfs.get('Group 1', pd.DataFrame())
group_2_df = grouped_dfs.get('Group 2', pd.DataFrame())
group_3_df = grouped_dfs.get('Group 3', pd.DataFrame())
```

```python
group_1_df.head(10)
```

```python
# Save the grouped data

group_1_df.to_excel('group_1_df.xlsx', index = False)
group_2_df.to_excel('group_2_df.xlsx', index = False)
group_3_df.to_excel('group_3_df.xlsx', index = False)
```

# Group Calculation

```python
# Pick rows needed for time calculation

time_1 = group_1_df[['CCSJ', 'BID DAYS', 'CO DAYS', 'DAYS ALLOWED']]
time_2 = group_2_df[['CCSJ', 'BID DAYS', 'CO DAYS', 'DAYS ALLOWED']]
time_3 = group_3_df[['CCSJ', 'BID DAYS', 'CO DAYS', 'DAYS ALLOWED']]
```

```python
time_1.head(10)
```

```python
# Save the data

time_1.to_excel('time_1.xlsx', index = False)
time_2.to_excel('time_2.xlsx', index = False)
time_3.to_excel('time_3.xlsx', index = False)
group_1_data.to_excel('group1.xlsx', index = False)
group_2_data.to_excel('group2.xlsx', index = False)
group_3_data.to_excel('group3.xlsx', index = False)
```

```python
# Merge the two dataframes based on the 'CCSJ' column

merged_time1 = pd.merge(group_1_data, time_1, on='CCSJ', how='left')
merged_time2 = pd.merge(group_2_data, time_2, on='CCSJ', how='left')
merged_time3 = pd.merge(group_3_data, time_3, on='CCSJ', how='left')
```

```python
# Check data

merged_time1.head(10)
```

```python
# Calculate the two time percentages for groups

merged_time1['Time_complete_OCD'] = merged_time1['Cumulative Charge Days'] / merged_time1['BID DAYS']
merged_time1['Time_complete_RCD'] = merged_time1['Cumulative Charge Days'] / merged_time1['DAYS ALLOWED']

merged_time2['Time_complete_OCD'] = merged_time2['Cumulative Charge Days'] / merged_time2['BID DAYS']
merged_time2['Time_complete_RCD'] = merged_time2['Cumulative Charge Days'] / merged_time2['DAYS ALLOWED']

merged_time3['Time_complete_OCD'] = merged_time3['Cumulative Charge Days'] / merged_time3['BID DAYS']
merged_time3['Time_complete_RCD'] = merged_time3['Cumulative Charge Days'] / merged_time3['DAYS ALLOWED']
```
